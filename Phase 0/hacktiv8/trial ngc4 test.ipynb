```python
class Book:
    #mendefinisikan kelas Book
    def __init__(self,title, author, book_id):
    #meninisiasi title, author dan ref id dari buku
        self.title = title
        self.author = author
        self.book_id = book_id
    def __str__(self):
    #menampilkan judul, penulis dan id dari buku
        return f"Title : {self.title}, Author : {self.author}, ID : {self.book_id}" 

class LibCatalog:
    #membuat catalog untuk isi lib
    def __init__(self):
        self.catalog = {}

    #menambah buku
    def add_book(self, book):
        self.catalog[book.book_id] = book
        print(f"Book '{book.title}' added sucessfully.")

    #mengurangi buku
    def remove_book(self,book_id):
        if book_id in self.catalog:
            remove_book = self.catalog.pop(book_id)
            print(f"Book '{remove_book.title}' removed sucessfully.")
        else:
            print(f"Error : Book ID not found") 
    #cari buku berdasarkan judul
    def search_title(self,title):
        found_books = [book for book in self.catalog.values() if title.lower() in book.title.lower()]
        if found_books:
            print(f"Books found with title'{title}':")
            for book in found_books:
                print(book)
        else:
            print(f"No books found by title '{title}.")
    
    #cari buku berdasarkan penulis
    def search_author(self,author):
        found_books = [book for book in self.catalog.values() if author.lower() in book.author.lower()]
        if found_books:
            print(f"Books found by author '{author}':")
            for book in found_books:
                print(book)
        else:
            print(f"No books found by author '{author}.")
    
    #mengeluarkan semua isi buku
    def display_books(self):
        if self.catalog:
            print("Current books in the catalog:")
            for book in self.catalog.values():
                print(book)
        else:
            print("The catalog is empty.")  
```

```python
import unittest

class TestLibraryCatalog(unittest.TestCase):
    
    def setUp(self):
        self.library = LibCatalog()
        self.book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", 0)
        self.book2 = Book("1984", "George Orwell", 1)
        self.book3 = Book("To Kill a Mockingbird", "Harper Lee", 2)

        self.library.add_book(self.book1)
        self.library.add_book(self.book2)
        self.library.add_book(self.book3)

    def test_add_book(self):
        new_book = Book("Brave New World", "Aldous Huxley", 4)
        self.library.add_book(new_book)
        self.assertIn(new_book.book_id, self.library.catalog)
        self.assertEqual(self.library.catalog[4].title, "Brave New World")

    def test_remove_book(self):
        self.library.remove_book(1)
        self.assertNotIn(1, self.library.catalog)

    def test_search_by_title(self):
        """Test searching books by title."""
        result = self.library.search_title("1984")
        self.assertEqual(len(result), 1)
        self.assertEqual(result[0].title, "1984")

    def test_search_by_author(self):
        """Test searching books by author."""
        result = self.library.search_author("Harper Lee")
        self.assertEqual(len(result), 1)
        self.assertEqual(result[0].author, "Harper Lee")

    def test_display_books(self):
        """Test displaying books in the catalog."""
        with self.assertLogs(level='INFO') as log:
            self.library.display_books()
            self.assertIn("The Great Gatsby", log.output[0])
            self.assertIn("1984", log.output[1])
            self.assertIn("To Kill a Mockingbird", log.output[2])

unittest.main(argv=[''], verbosity=2, exit=False)
```
